### Unconfined Concrete

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fco = 5000 #psi
eco = (fco**0.25)/4000
Ec = 185000*(fco**(3/8)) #psi
r = (fco/750) - 1.9
strain = np.arange(0,0.0031,0.0001)
x = strain/eco
n = Ec*(eco/fco)
y = (n*x)/(1 + ( n - r/(r-1) )*x + (x**r)/(r-1))
stress = fco*y
```

```python
plt.plot(strain,stress)
plt.grid(True, alpha = 0.3)
plt.xlabel('Compressive strain (in/in)')
plt.ylabel('Compressive stress (psi)')
plt.xlim([0, 0.003])
plt.ylim([0, 6000])
plt.title('Stress-strain unconfined concrete')
plt.show()
```
